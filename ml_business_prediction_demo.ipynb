## Machine Learning Concept

**Past Data → Pattern → Prediction**

Machine Learning studies past data to find patterns and make predictions.

```python
# Import ML model
from sklearn.linear_model import LinearRegression
```

## Prepare Data for Machine Learning

X → Input variable

y → Target variable

```python
X = df[['marketing_spend']]
y = df['sales']
```

## Train the Model

The model learns the relationship between marketing spend and sales.

```python
model = LinearRegression()
model.fit(X, y)
```

## Make Predictions

Now we ask the model:

**If marketing spend is ₹6000, what will be the predicted sales?**

```python
prediction = model.predict([[6000]])
prediction
```

## Visualize the Prediction Line

The line represents the trend learned by the Machine Learning model.

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X),color="red")
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.title('Machine Learning Prediction')
plt.show()
```
